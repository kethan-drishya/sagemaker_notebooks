```python

df['worktype']= df['worktype'].astype(str)
df['description']= df['description'].astype(str)
df['location']= df['location'].astype(str)
df['failurecode']= df['failurecode'].astype(str)
df['siteid']= df['siteid'].astype(str)
df['craft']= df['craft'].astype(str)
df['approve_type']= df['approve_type'].astype(str)
```

```python
def asset_analysis(dataframe,cat):
    fig, ax = plt.subplots()
    #color=['sandybrown', 'lightskyblue', 'lightskyblue', 'lightskyblue','peachpuff', 'steelblue']
    my_bar=dataframe[cat].value_counts()[:10].plot(kind='bar', figsize=(20,4),ax=ax);
    
    sns.set(font_scale=1.04)
    plt.xlabel(cat, fontsize=10,fontweight='bold')
    plt.ylabel("Count of failures", fontsize=25,fontweight='bold')
    plt.title("bar plot",fontsize=25,fontweight='bold');
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),fontsize=25, va='bottom')
    import matplotlib 
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
    ax.set_facecolor('xkcd:white')
    fig.patch.set_facecolor('xkcd:white')
    matplotlib.rc('xtick', labelsize=25) 
    matplotlib.rc('ytick', labelsize=25)
```

```python
asset_analysis(df,'description')
```

```python
asset_analysis(df[df['worktype']!='PM'],'description')
```
