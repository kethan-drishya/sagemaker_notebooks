```python
def get_col_names_dict(df_tag, col_names):
    match_dict = {}
    i = 0   
    for key in df_tag.keys():
        match_dict[col_names[i]] = key
        i = i + 1
    return match_dict 
col_names = ['hardness', 'iron', 'silica', 'ph', 'conductivity', 'turbidity','P1','EDTA', 'CAU']
match_dict_chem = get_col_names_dict(df_tag_chem, col_names)
print(match_dict_chem)
```

```python
import matplotlib.pyplot as plt
def plot_hist(df, tag, title):
    plt.hist(df[tag], bins=50, alpha=0.5,
                    range=[df[tag].min(), df[tag].max()],
                      facecolor='gray', align='mid')
    plt.title(title, fontsize=10, fontweight='bold')
    plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


def plot_elbow(K, values, ylabel):
    plt.plot(K, values, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel(ylabel)
    plt.title(f"The Elbow Method using {ylabel}")
    plt.show()
    
def kmeans_func(df_td_1d):
    columns = ['hardness', 'iron', 'silica', 'ph', 'conductivity', 'turbidity', 'U1']
    X = df_td_1d[columns]
    normalizer = StandardScaler()
    X_norm = normalizer.fit_transform(X.values)

    # X_train, X_test = train_test_split(X, test_size = 0.25, shuffle=False)
    # normalizerx.fit(X_train.values)
    # X_train_norm = normalizerx.transform(X_train.values)
    # X_test_norm = normalizerx.transform(X_test.values)


    distortions = []
    inertias = []
    ss = []

    K = range(2, 100, 5)

    for k in K:
        # Building and fitting the model
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(X_norm)

        distortions.append(sum(np.min(cdist(X_norm, kmeanModel.cluster_centers_,
                                            'euclidean'), axis=1)) / X.shape[0])
        inertias.append(kmeanModel.inertia_)
        ss.append(silhouette_score(X_norm, kmeanModel.labels_))
 
    
    

    plot_elbow(K, distortions, 'Distortion')
    plot_elbow(K, inertias, 'Inertia')
    plot_elbow(K, ss, 'silhouette')

    



```

```python
from scipy.stats import anderson
# normality test
for col in df_td.columns:
    result = anderson(df_td[col].values)
    print('Statistic: %.3f' % result.statistic)
    p = 0
    for i in range(len(result.critical_values)):
        sl, cv = result.significance_level[i], result.critical_values[i]
        if result.statistic < result.critical_values[i]:
            print(f'%.3f: %.3f, {col} looks normal (fail to reject H0)' % (sl, cv))
        else:
            print(f'%.3f: %.3f, {col} does not look normal (reject H0)' % (sl, cv))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor 
def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```
