```python
df['worktype']= 
df['description']=
df['location']=
df['failurecode']=
df['siteid']=
df['craft']=
df['approve_type']=
```

```python
diff_list1=[]
diff_list2=[]
diff_list3=[]
for ind in df.index: 
    d1 = df['actfinish'][ind] - df['actstart'][ind]
    d2 = df['actstart'][ind] - df['reportdate'][ind]
    d3 = df['actfinish'][ind] - df['reportdate'][ind]
    diff_list1.append(d1)
    diff_list2.append(d2)
    diff_list3.append(d3)
df['act_diff'] = diff_list1 
df['report_diff'] = diff_list2   
df['complete_diff'] = diff_list3
```

```python
df['seconds1']=df['act_diff'].astype('timedelta64[s]')
df['seconds2']=df['report_diff'].astype('timedelta64[s]')
df['seconds3']=df['complete_diff'].astype('timedelta64[s]')
df1=df.sort_values(by=['seconds1'],ascending=False)
df2=df.sort_values(by=['seconds2'],ascending=False)
df3=df.sort_values(by=['seconds3'],ascending=False)
```

```python
def asset_analysis(dataframe,cat):
    fig, ax = plt.subplots()
    #color=['sandybrown', 'lightskyblue', 'lightskyblue', 'lightskyblue','peachpuff', 'steelblue']
    my_bar=dataframe[cat].value_counts()[:10].plot(kind='bar', figsize=(20,4),ax=ax);
    
    sns.set(font_scale=1.04)
    plt.xlabel(cat, fontsize=25,fontweight='bold')
    plt.ylabel("Count of failures", fontsize=25,fontweight='bold')
    plt.title("bar plot",fontsize=25,fontweight='bold');
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005),fontsize=25, va='bottom')
    import matplotlib 
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
    ax.set_facecolor('xkcd:white')
    fig.patch.set_facecolor('xkcd:white')
    matplotlib.rc('xtick', labelsize=25) 
    matplotlib.rc('ytick', labelsize=25)
```

```python
asset_analysis(df111,'description')
```
